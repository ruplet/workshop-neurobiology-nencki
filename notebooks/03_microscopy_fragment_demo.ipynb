# Prosty demo obrazów mikroskopowych (warsztat 2018)

To tylko demonstracja materiału warsztatowego.

```python
%matplotlib inline
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import tifffile
```

```python
participant = 'participant_01'
base = Path('../data/microscopy') / participant
dapi = tifffile.imread(base / f'{participant}_DAPI.tif')
ch488 = tifffile.imread(base / f'{participant}_488.tif')

def norm01(img):
    lo, hi = np.percentile(img, [1, 99.5])
    return np.clip((img.astype(np.float32) - lo) / (hi - lo), 0, 1)

dapi_n = norm01(dapi)
ch488_n = norm01(ch488)
rgb = np.dstack([np.zeros_like(dapi_n), ch488_n, dapi_n])
```

```python
plt.figure(figsize=(9, 7))
plt.imshow(rgb)
plt.title('Pełny obraz: 488 (zielony) + DAPI (niebieski)')
plt.axis('off')
```

```python
x, y, w, h = 512, 384, 256, 256
frag = rgb[y:y+h, x:x+w]

plt.figure(figsize=(6, 6))
plt.imshow(frag)
plt.title('Fragment obrazu (demo)')
plt.axis('off')
```

Simple note: this demo shows DAPI and 488 dyes. Here, `DAPI` and `488` are just names of dyes/channels from the original file naming.
